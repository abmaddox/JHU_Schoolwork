```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score


#This is a modified version of my code from assignment 4

def ScoreModels(_xTrain, _yTrain, _xtest, _ytest, _param_grid_dict):
    #Stores the best models from each trial of parameter tuning
    bestModels = {}
    scores = []
    
    # Tunes hyperparameters for each model, 
    #gets F1 score for each model type, and stores the best perfroming model in a dictionary
    for key, value in _param_grid_dict.items():
        
        pipe = make_pipeline(StandardScaler(), value['model'])
        gs = GridSearchCV(estimator = pipe, 
                          param_grid= value['params'],
                          n_jobs = -1,
                          cv = 10,
                          scoring='f1')
        
        
        gsFit = gs.fit(_xTrain, _yTrain)
        
        bestModels.update({key : {
            'model':gsFit.best_estimator_,
            'score':f1_score(_ytest, gsFit.predict(_xtest) )
        } })
        
        print(key, "completed.")
    return bestModels
```

```python
import torch
from torch import nn
from torch import optim

# Convert to tensors
X_train, Y_train = torch.tensor(X_train.to_numpy(dtype=np.float32)), torch.tensor(Y_train.to_numpy(dtype=np.int64))
X_test, Y_test = torch.tensor(X_test.to_numpy(dtype=np.float32)), torch.tensor(Y_test.to_numpy(dtype=np.int64))
```

```python
print(Y_train.shape)
print(X_train.shape)
```

```python
class OneLayerNN(nn.Module):
    """ A PyTorch neural network model """
    def __init__(self, n_hidden=30, epochs=100, eta=0.05, minibatch_size=100):
        super(OneLayerNN, self).__init__()
        self.n_hidden = n_hidden  # hidden layer size
        self.epochs = epochs  # number of learning iterations
        self.eta = eta  # learning rate
        self.minibatch_size = minibatch_size  # size of training batch - 1 would not work
        self.fc1, self.fc2, self.fc3 = None, None, None

    def _forward(self, X, apply_softmax=False):
        assert self.fc1 != None
        X = nn.functional.relu(self.fc1(X))
        X = nn.functional.relu(self.fc2(X))
        X = self.fc3(X)
        if apply_softmax:
            X = nn.functional.softmax(X, dim=1)
        return X

    def _reset(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                 m.reset_parameters()

    def predict(self, _X):
        assert self.fc1 is not None
        net_out = self._forward(_X, apply_softmax=True)
        p_values, indices = net_out.max(dim=1)
        return indices

    def fit(self, X_train, y_train):
        self._reset()  # Reset the neural network weights
        n_output= np.unique(y_train).shape[0]  # number of class labels
        n_features= X_train.shape[1]

        self.fc1 = nn.Linear(n_features, self.n_hidden)  # A simple input layer
        self.fc2 = nn.Linear(self.n_hidden, self.n_hidden)  # A simple hidden layer
        self.fc3 = nn.Linear(self.n_hidden, n_output)  # A simple output layer
        
        optimizer = optim.SGD(self.parameters(), lr=self.eta, momentum=0.9)
        loss_func = nn.CrossEntropyLoss()# only two classes in this case

        for _ in range(self.epochs):
            indices = np.arange(X_train.shape[0])
            
            for start_idx in range(0, indices.shape[0] - self.minibatch_size + 1, self.minibatch_size):
                batch_idx = indices[start_idx:start_idx + self.minibatch_size]
                optimizer.zero_grad()
                
                net_out = self._forward(X_train[batch_idx])

                loss = loss_func(net_out, y_train[batch_idx])
                loss.backward()
                optimizer.step()
```

```python
class TwoLayerNN(nn.Module):
    """ A PyTorch neural network model """
    def __init__(self, n_hidden=30, epochs=100, eta=0.05, minibatch_size=100):
        super(TwoLayerNN, self).__init__()
        self.n_hidden = n_hidden  # hidden layer size
        self.epochs = epochs  # number of learning iterations
        self.eta = eta  # learning rate
        self.minibatch_size = minibatch_size  # size of training batch - 1 would not work
        self.fc1, self.fc2, self.fc3, self.fc4 = None, None, None, None # now has two hidden layers

    def _forward(self, X, apply_softmax=False):
        assert self.fc1 != None
        X = nn.functional.relu(self.fc1(X))
        X = nn.functional.relu(self.fc2(X))
        X = nn.functional.relu(self.fc3(X)) # modified for the second hidden layer
        X = self.fc4(X)
        if apply_softmax:
            X = nn.functional.softmax(X, dim=1)
        return X

    def _reset(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                 m.reset_parameters()

    def predict(self, _X):
        assert self.fc1 is not None
        net_out = self._forward(_X, apply_softmax=True)
        p_values, indices = net_out.max(dim=1)
        return indices

    def fit(self, X_train, y_train):
        self._reset()  # Reset the neural network weights
        n_output= np.unique(y_train).shape[0]  # number of class labels, two in this case
        n_features= X_train.shape[1]

        self.fc1 = nn.Linear(n_features, self.n_hidden)  # A simple input layer
        self.fc2 = nn.Linear(self.n_hidden, self.n_hidden)  # A simple hidden layer
        self.fc3 = nn.Linear(self.n_hidden, self.n_hidden)  # A simple hidden layer
        self.fc4 = nn.Linear(self.n_hidden, n_output)  # A simple output layer
        
        optimizer = optim.SGD(self.parameters(), lr=self.eta, momentum=0.9)
        loss_func = nn.CrossEntropyLoss()

        for _ in range(self.epochs):
            indices = np.arange(X_train.shape[0])
            
            for start_idx in range(0, indices.shape[0] - self.minibatch_size + 1, self.minibatch_size):
                batch_idx = indices[start_idx:start_idx + self.minibatch_size]
                optimizer.zero_grad()    
                
                net_out = self._forward(X_train[batch_idx])

                loss = loss_func(net_out, y_train[batch_idx])
                loss.backward()
                optimizer.step()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
    
def kfold_eval_acc(_clf, _X, _y):
    

    # Need indexable data structure
    accuracy = []
    kf = StratifiedKFold(n_splits=10, shuffle=False, random_state=None)
    for train_index, test_index in kf.split(_X, _y):
        _clf.fit(_X[train_index], _y[train_index])
        ypred = _clf.predict(_X[test_index])
        accuracy += [accuracy_score(_y[test_index], ypred)]
    return np.array(accuracy)
```

```python
def kfold_eval_f1(_clf, _X, _y):
    from sklearn.model_selection import StratifiedKFold
    from sklearn.metrics import confusion_matrix

    kf = StratifiedKFold(n_splits=10, shuffle=False, random_state=None)
    
    fn = 0
    fp = 0
    tp = 0
    tn = 0
    
    for train_index, test_index in kf.split(_X, _y):
        _clf.fit(_X[train_index], _y[train_index])
        ypred = _clf.predict(_X[test_index])
        
        for truth, pred in zip(_y[test_index], ypred):
            if (truth == 0 and pred == 0):
                tn += 1
            elif (truth == 0 and pred == 1):
                fp += 1
            elif (truth == 1 and pred == 0):
                fn += 1
            elif (truth == 1 and pred == 1):
                tp += 1
        print(f"fp: {fp}  fn:  {fn}  tp:  {tp}  tn:  {tn}")
    
    if((tp == 0) or (fn == 0) or (fp == 0)):            
        return _clf, 0
    else:
        prec = tp/(tp+fp)
        recall = tp/(tp+fn)
        f1 = 2*(prec*recall)/(prec+recall)
        return _clf, f1
```

```python
%%time
# runs in 30 minutes
clf = OneLayerNN(n_hidden = 40, epochs = 200, eta = 0.07, minibatch_size=200)

fit_clf, f1 = kfold_eval_f1(clf, X_train, Y_train)

print(f"PyTorch Feedforward NN 10-fold CV f1 = {f1:.4f}")
nnOnePred = fit_clf.predict(X_test)
```

```python
print(f"PyTorch Feedforward NN test f1 = {f1_score(Y_test, nnOnePred):.4f}")
print(f"PyTorch Feedforward NN test accuracy = {accuracy_score(Y_test, nnOnePred):.4f}")
```

```python
class OneLayerDropoutNN(nn.Module):
    """ A PyTorch neural network model """
    def __init__(self, n_hidden=30, epochs=100, eta=0.05, minibatch_size=100, apply_dropout = False, drop_prob = 0.5):
        super(OneLayerDropoutNN, self).__init__()
        self.n_hidden = n_hidden  # hidden layer size
        self.epochs = epochs  # number of learning iterations
        self.eta = eta  # learning rate
        self.minibatch_size = minibatch_size  # size of training batch - 1 would not work
        self.fc1, self.fc2, self.fc3 = None, None, None
        self.p = drop_prob
        self.dropout = apply_dropout

    def _forward(self, X, apply_softmax=False, apply_dropout=False):
        assert self.fc1 != None
        if apply_dropout:
            d = nn.Dropout(p = self.p) #probability that a node is dropped
            X = d(X)
            
        X = nn.functional.relu(self.fc1(X))
        
        if apply_dropout:
            X = d(X)
            
        X = nn.functional.relu(self.fc2(X))
        X = self.fc3(X)
        
        if apply_softmax:
            X = nn.functional.softmax(X, dim=1)
        return X

    def _reset(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                 m.reset_parameters()

    def predict(self, _X):
        assert self.fc1 is not None
        net_out = self._forward(_X, apply_softmax=True)
        p_values, indices = net_out.max(dim=1)
        return indices

    def fit(self, X_train, y_train):
        self._reset()  # Reset the neural network weights
        n_output= np.unique(y_train).shape[0]  # number of class labels
        n_features= X_train.shape[1]

        self.fc1 = nn.Linear(n_features, self.n_hidden)  # A simple input layer
        self.fc2 = nn.Linear(self.n_hidden, self.n_hidden)  # A simple hidden layer
        self.fc3 = nn.Linear(self.n_hidden, n_output)  # A simple output layer
        
        optimizer = optim.SGD(self.parameters(), lr=self.eta, momentum=0.9)
        loss_func = nn.CrossEntropyLoss()# only two classes in this case

        for _ in range(self.epochs):
            indices = np.arange(X_train.shape[0])
            
            for start_idx in range(0, indices.shape[0] - self.minibatch_size + 1, self.minibatch_size):
                batch_idx = indices[start_idx:start_idx + self.minibatch_size]
                optimizer.zero_grad()
                
                net_out = self._forward(X_train[batch_idx], apply_dropout=self.dropout)

                loss = loss_func(net_out, y_train[batch_idx])
                loss.backward()
                optimizer.step()
```

```python
%%time
#runs in 27 mins
clf3 = OneLayerDropoutNN(n_hidden = 40, epochs = 200, eta = 0.07, minibatch_size=200, apply_dropout = True, drop_prob = .6)

fit_clf3, f1_3 = kfold_eval_f1(clf3, X_train, Y_train)

print(f"PyTorch Feedforward NN with 40% dropout 10-fold CV f1= {f1_3:.4f}")

nnOne40DropPred = fit_clf3.predict(X_test)

```

```python
print(f"PyTorch Feedforward NN 1 hidden layer 40% dropout test f1 = {f1_score(Y_test, nnOne40DropPred):.4f}")
print(f"PyTorch Feedforward NN 1 hidden layer 40% dropout test accuracy = {accuracy_score(Y_test, nnOne40DropPred):.4f}")
```

A more robust model is preferred because its performance is affected less by unseen data. It is less overfit and less reliant upon singular weights, in the NN model case, and forms new relationships with other nodes to form the same conlcusions. This makes the model more generalizable.
