```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display

#Importing the dataset
mushData = "./Datasets/mushroom_dataset.csv"

mushDF1 = pd.read_csv(mushData)

display(mushDF1)
```

```python
#Changing the nominal classes to numerical
class_map = {'p': 0,
            'e': 1}
mushDF1['class'] = mushDF1['class'].map(class_map)

display(mushDF1)
```

```python
#Checking for nulls
display(mushDF1.isnull().any())
```

```python
# Separating the targets from the data
targets = mushDF1['class']
mushDF1.drop(columns=['class'], inplace=True)

display(targets)
display(mushDF1)
```

```python
# Unknown if the features are ordinal or the relationship of their ordinal values to nominal, so they must be one hot encoded
mushDF2 = pd.get_dummies(mushDF1)

display(mushDF2)
```

```python
# PLotting values to check distribution
fig, axs = plt.subplots(nrows=10, ncols=12, sharex=True, figsize = (20, 10), tight_layout = True)
Ys = targets.ravel()
    
for num, ax in enumerate(axs.flat):
    if num < len( mushDF2.columns ):
        Xs = mushDF2[mushDF2.columns[num]].values
        ax.scatter(Xs, Ys)
        ax.set_title(mushDF2.columns[num], fontsize = 'small')
    else:
        ax.axis('off')
    
```

## Question 1

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

#Dataset splitting
xTrain, xTest, yTrain, yTest = train_test_split(mushDF2, targets.ravel(),
                                                train_size= .8, stratify = targets.ravel(),
                                                random_state=5)

#Checking pipeline with linearSVC
lsvc = LinearSVC()
lsvc.fit(xTrain, yTrain)
print(lsvc.score(xTest, yTest))
```

```python
#Testing all models with 10 fold cv
clfs = [MLPClassifier(), GaussianNB(), SVC(kernel = 'linear', probability= True), DecisionTreeClassifier()]

for clf in clfs:
    scores = cross_val_score(clf, xTrain, yTrain, cv=10, n_jobs=-1)
    print(f'{clf}\t Accuracy: {scores.mean() : .2f}')
```

```python
RF = RandomForestClassifier()
scores = cross_val_score(RF, xTrain, yTrain, cv=10, n_jobs=-1)
print(f'{RF}\t Accuracy: {scores.mean() : .2f}')
```

The dataset is easily classified because the clusters are very tight and non-overlapping. All models except for the GaussianNB achieved 100% accuracy in 10 fold cross validation. It makes me suspicious that I did something incorrectly, but I cannot figure out what it could be. 

## Question 2

```python
ensemble_clfs = []

#New weak versions of the classifiers
weak_clfs = [MLPClassifier(hidden_layer_sizes=(3,3), max_iter=30, tol=1e-1),
             GaussianNB(),
             SVC(kernel = 'linear', probability= True),
             DecisionTreeClassifier(max_depth=5, max_features=5)]

#Instantiating ensemble array
for clf in  weak_clfs:
    weak = [clf for i in range(100)]
    ensemble_clfs.append(weak)

display(ensemble_clfs)
```

```python
for row in range(len(ensemble_clfs)):
    weakling = ensemble_clfs[row][0].fit(xTrain, yTrain)
    scores = cross_val_score(weakling, xTrain, yTrain, cv=10, n_jobs=-1)
    print(f'{weakling}\t Accuracy: {scores.mean() : .2f}')
```

Despite weakening the classifiers they still perform well. My suspicion that I've made a grave error grows.
